### 실습 01 (타이타닉 생존자 체크)

```python
import pandas as pd
data = pd.read_csv('titanic_train.csv')
data
```

```python
def Generation(n):
    if n // 10 == 0:
        return "유아"
    elif n // 10 == 1:
        return "10대"
    elif n // 10 == 2:
        return "20대"
    elif n // 10 == 3:
        return "30대"
    elif n // 10 == 4:
        return "40대"
    elif n // 10 == 5:
        return "50대"
    elif n // 10 == 6:
        return "60대"
    else:
        return "노인"
    
data["Age_Range"] = data[data["Survived"] == 1]["Age"].apply(Generation)
data[data["Survived"] == 1]["Age_Range"].value_counts().sort_index()
```

```python
plt.bar(['10대', '20대', '30대', '40대', '50대', '60대', '노인', '유아'],
        data[data["Survived"] == 1]["Age_Range"].value_counts().sort_index(), color = 'g')
plt.title('나이대별 타이타닉 생존자')
plt.xlabel('나이')
plt.ylabel('생존자')
plt.show()
```

```python
plt.bar(['female', 'male'],
       data[data["Survived"] == 1]["Sex"].value_counts(), color = 'g')
plt.title('나이대별 타이타닉 생존자')
plt.xlabel('성별')
plt.ylabel('생존자')
plt.show()
```

```python
data = pd.read_csv('kr_daily.csv')
data
```

```python
str(data['date'][0])[:4] + "-" + str(data['date'][0])[4:6] + "-" + str(data['date'][0])[6:8]
```

```python
len(data['date'])
```

```python
data['date_format'] = None
for i in range(len(data['date'])):
    data['date_format'][i] = str(data['date'][i])[:4] + "-" + str(data['date'][i])[4:6] + "-" + str(data['date'][i])[6:8]
    
data
```

```python
plt.plot(data['date_format'][118:], data['confirmed'][118:], linestyle='--', color = 'r')
plt.xticks(rotation=60)
plt.title('확진자')
plt.show()
```

```python
plt.bar(data['date_format'][118:], data['death'][118:], color = 'r')
plt.xticks(rotation=60)
plt.title('사망자')
plt.show()
```
